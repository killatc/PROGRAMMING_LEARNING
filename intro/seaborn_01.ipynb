```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
a = np.random.randint(1,101,10)
b = np.random.randint(1,101,10)

x = np.arange(0,10)
```

```python
fig,ax =plt.subplots(ncols=1,nrows=1,figsize =(9,5))
sns.lineplot(x=x,y=a,color ='red', marker='d',linestyle='dashed', linewidth=2,markersize=10,label='red line')

sns.lineplot(x=x,y=b,color='purple',marker='*', markerfacecolor='#be95c9',
              markersize=12, label='purple line' ).set(title ='Plot with seaborn',xlabel='Position', ylabel='Random number')


# ax.set_ylabel('Random number')
ax.legend(loc ='lower left')
ax.grid(ls = '--',lw =0.8,alpha =0.4)


ax2=ax.twinx().twiny()

sns.lineplot(x=x,y=a,alpha =0,ax=ax2)
sns.lineplot(x=x,y=b,alpha =0,ax=ax2)


plt.savefig('./Plots/sns_lineplot.pdf')
plt.savefig('./Plots/sns_lineplot.png', dpi =300, pad_inches = 0.1,bbox_inches = 'tight')

```

```python
dist_normal = np.random.normal(0,100,1000) 
dist_exp = np.random.exponential(100,1000)
```

```python
fig,ax =plt.subplots(ncols=2,nrows=1,figsize =(12,5))
sns.histplot(dist_normal,kde=True, ax = ax[0], color = 'red')
sns.histplot(dist_exp, ax = ax[1], kde =True)
ax[0].set_title('Normal distribution')
ax[1].set_title('Exponential distribution')
ax[0].set_xlabel('Numbers')
ax[0].set_ylabel('Frequency')
ax[1].set_xlabel('Numbers')
ax[1].set_ylabel('Frequency')

#hello, my love <3 :P
```

```python
# subplots as a list grid
fig, ax = plt.subplots(ncols=3, nrows=1, figsize=(9,5))

print(type(ax)), print(ax.shape)
ax[1].plot([1, 2, 3])
```

```python
# subplots as a matrix grid
fig, ax = plt.subplots(ncols=3, nrows=2, figsize=(9,5))

print(type(ax)), print(ax.shape)
ax[0][1].plot([1, 2, 3])
ax[0][2].plot([1, 2, 5])
ax[0][0].set_title("1st line, 1st column")
ax[0][1].set_title('1st line, 2nd column')
ax[0][2].set_title('1st line, 3rd column')
ax[1][0].set_title('2nd line, 1st column')
ax[1][1].set_title('2nd line, 2nd column')
ax[1][2].set_title('2nd line, 3rd column')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.2,
                    hspace=0.4)
```

```python
fig,ax = plt.subplots(ncols=2,nrows=2, figsize =(15,8))
dist_normal = np.random.normal(10,100,2000)
bins_normal = np.arange(-300,350,50)
dist_unifor = np.random.uniform(10,2000,2500)
dist_expo   = np.random.exponential(2500,2000)
dist_lapla  = np.random.laplace(10,100,2000)

#print(bins_normal)

sns.histplot(dist_normal,ax =ax[0][0],kde =True,stat='count', bins=bins_normal,color = 'purple'
).set(
    title ='Normal Distribution',
    xlabel='Random numbers',
    ylabel='Absolute Frenquency'
) 
# kde = function the smooth
# stat = propability(normalizar) or count()
#bins=[-300,-100,100,300], bins= 10

sns.histplot(dist_unifor,ax =ax[0][1],stat='probability',color ='lightblue'
).set(
    title ='Uniforme Distribution ',
    xlabel='Random numbers',
    ylabel='Relative Frenquency'
) 

ax2 = ax[0][1].twinx()
sns.kdeplot(dist_unifor, ax =ax2, color ='red') 


sns.histplot(
    dist_expo,
    ax =ax[1][0],
    kde =True,
    stat='probability',color = 'red'
).set(
    title ='Exponetial Distribution',
    xlabel='Random numbers',
    ylabel='Relative Frenquency'
)

# Plotting laplace distribution using seaborn
sns.histplot(
    dist_lapla,
    ax=ax[1][1],
    kde =True, 
    stat='count'
).set(
    title ='Laplace Distribution',
    xlabel='Random numbers', 
    ylabel='Absolute Frenquency'
) 



plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.2,
                    hspace=0.4)
```
