```python
a = np.random.randint(1,101,10)
b = np.random.randint(1,101,10)

x = np.arange(0,len(a))

```

```python

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (7,5))

ax.plot(x,a, linewidth = 1.5,linestyle= '--',color = 'b',alpha = 0.5, label='blue line')
ax.plot(x,b, color = 'orange',linestyle= 'dashdot',label= 'orange line')
ax.set_title('This is a Title',fontsize =10,fontweight ='bold')
ax.set_xlabel('Example of xlabel',fontsize =20)
ax.set_ylabel('Example of ylabel',fontsize =50)
ax.legend(title = 'legend exmple',fontsize =5)
ax.grid(ls = '--',lw =0.8,alpha =0.5)
```

```python
fig, ax =plt.subplots(nrows =1, ncols =1,figsize = (10,5))
ax.scatter(x,a, color ='lightblue', s = 120, marker = '^',label ='ligthblue triangle')
ax.plot(x,a, color ='purple', linewidth = 1.8, linestyle = '--', alpha = 0.5)

ax.scatter(x,b, color ='gold', s = 150, marker = 'o',alpha =1,label= 'gold circle')
ax.plot(x,b, color = 'orange',linestyle= 'dashdot')


ax.grid(ls = ':', alpha =0.9,lw = 0.8)
ax.legend()
ax.set_title('This is a Title is an example',fontsize =15,fontweight ='bold')
ax.set_ylabel('Example the scatterplot (ylabel)',fontsize =15)
ax.set_xlabel('Example the scatterplot (xlabel)',fontsize =15)
```

```python

fig,ax = plt.subplots(ncols= 2, nrows=1,figsize=(10,4),sharey=True)
ax[0].bar(x,a, color='#3A86FF')
ax[1].bar(x,b,color ='#ffafcc')
ax[0].set_title('Radom Numbers of A',fontsize = 16,fontweight ='bold', color ='lightblue')
ax[1].set_title('Random Numbers of B', fontsize = 16,fontweight ='bold',color = '#ffafcc')
ax[0].set_ylabel('Random Numbers from 0 to 100', fontsize =12)
ax[1].grid(ls = '--', alpha =0.5,lw = 0.5)
```

```python
fig,ax = plt.subplots(ncols=2, nrows=1,figsize=(10,4))
bp1 = ax[0].boxplot(a,patch_artist=True)
bp2 = ax[1].boxplot(b,patch_artist=True)

bp1['boxes'][0].set_facecolor('#cdb4db')
bp1['boxes'][0].set_linewidth(2)
bp1['boxes'][0].set_edgecolor('#00b4d8')

bp1['whiskers'][0].set_linewidth(2)
bp1['whiskers'][1].set_linewidth(2)
bp1['whiskers'][0].set_color('#fb6f92')
bp1['whiskers'][1].set_color('#02c39a')


bp1['caps'][0].set_color('#00b4d8')
bp1['caps'][1].set_color('#c9184a')

bp1['medians'][0].set_color('#725ac1')
bp1['medians'][0].set_linewidth(2)


bp2['boxes'][0].set_facecolor('#ffe5ec')
bp2['boxes'][0].set_linewidth(2)
bp2['boxes'][0].set_edgecolor('#fb6f92')

bp2['whiskers'][0].set_linewidth(2)
bp2['whiskers'][1].set_linewidth(2)
bp2['whiskers'][0].set_color('#00b4d8')
bp2['whiskers'][1].set_color('#02c39a')


bp2['caps'][0].set_color('#c9184a')
bp2['caps'][1].set_color('#c9184a')
bp2['caps'][1].set_linewidth(2)
bp2['caps'][0].set_linewidth(2)


bp2['medians'][0].set_color('#8ac926')
bp2['medians'][0].set_linewidth(2)


ax[0].set_title('Radom Numbers of A',fontsize = 16,fontweight ='bold', color ='lightblue')
ax[1].set_title('Random Numbers of B', fontsize = 16,fontweight ='bold',color = '#ffafcc')
ax[0].set_ylabel('Random Numbers from 0 to 100', fontsize =12)
ax[1].set_ylabel('Random Numbers from 0 to 100', fontsize =12)

ax[0].grid(ls = ':', alpha =0.9,lw = 0.8)
ax[1].grid(ls = ':', alpha =0.9,lw = 0.8)
```
